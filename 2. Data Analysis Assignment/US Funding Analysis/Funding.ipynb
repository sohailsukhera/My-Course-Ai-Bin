```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\My Course Ai Bin\Week 6 Assignments\startup_growth_investment_data.csv')
df
```

```python
industry_counts = df['Industry'].value_counts()
industry_counts
```

```python
plt.figure(figsize=(10, 6))
industry_counts.plot(kind='bar')
plt.title('Number of Startups per Industry')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
avg_growth = df.groupby('Industry')['Growth Rate (%)'].mean().sort_values(ascending=False)
avg_growth
```

```python
plt.figure(figsize=(10, 6))
avg_growth.plot(kind='bar')
plt.title('Average Growth Rate by Industry')
plt.xlabel('Industry')
plt.ylabel('Average Growth Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
top_growth = df.nlargest(10, 'Growth Rate (%)')[
    ['Startup Name', 'Industry', 'Country', 'Growth Rate (%)']
]
top_growth
```

```python

top_value = df.nlargest(10, 'Valuation (USD)')[
    ['Startup Name', 'Industry', 'Country', 'Valuation (USD)']
]
top_value
```

```python
country_counts = df['Country'].value_counts()
country_counts
```

```python
country_growth = df.groupby('Country')['Growth Rate (%)'].mean().sort_values(ascending=False)
country_growth.head(10)
```

```python
plt.figure(figsize=(12, 6))
country_growth.head(10).plot(kind='bar')
plt.title('Top 10 Countries by Average Growth Rate')
plt.xlabel('Country')
plt.ylabel('Average Growth Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
rounds_count = df['Funding Rounds'].value_counts().sort_index()
rounds_count
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Funding Rounds'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribution of Funding Rounds')
plt.xlabel('Number of Funding Rounds')
plt.ylabel('Number of Startups')
plt.tight_layout()
```

```python
year_counts = df['Year Founded'].value_counts().sort_index()
print(year_counts)
```

```python
plt.figure(figsize=(14, 6))
year_counts.plot(kind='line', marker='o')
plt.title('Number of Startups Founded Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Startups')
plt.grid(True, alpha=0.3)
plt.tight_layout()
```
