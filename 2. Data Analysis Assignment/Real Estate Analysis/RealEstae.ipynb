# Data Cleaning 

```python
df_clean = df.dropna(subset=['price', 'city'])
print("After cleaning records:", len(df_clean))
```

# Top Cities by Listings

```python
city_listings = df_clean['city'].value_counts()
city_listings.head(10)
```

# Average Price by City (Top 10)

```python
avg_price_by_city = df_clean.groupby('city')['price'].mean().sort_values(ascending=False)
avg_price_by_city.head(10)
```

# Average Price by Bedrooms


```python
bedroom_avg = df_clean.groupby('bed')['price'].mean()
bedroom_avg
```

## Price Distribution (Histogram)


```python
plt.figure(figsize=(10, 6))
plt.hist(df_clean['price'].dropna(), bins=40, edgecolor='black')
plt.xlabel('Price (USD)')
plt.ylabel('Number of Properties')
plt.title('Distribution of Property Prices')
```

## Top 10 Most Expensive Cities (Bar Chart)


```python
top_10_cities = avg_price_by_city.head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_cities.index, top_10_cities.values)
plt.xlabel('City')
plt.ylabel('Average Price (USD)')
plt.title('Top 10 Most Expensive Cities in Puerto Rico')
plt.xticks(rotation=45, ha='right')
```

## Price Distribution by Bedrooms


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bed', y='price', data=df_clean)
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price (USD)')
plt.title('Price Distribution by Number of Bedrooms')
```

## House Size vs Price Relationship


```python
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['house_size'], df_clean['price'], alpha=0.5)
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (USD)')
plt.title('Relationship between House Size and Price')
```
