# Top 10 Restaurant Chains 


```python
top_chains = df['name'].value_counts().head(10)
print("Top 10 restaurant chains by number of locations:")
top_chains
```

```python
plt.figure(figsize=(12, 6))
top_chains.plot(kind='bar')
plt.xlabel('Restaurant Chain')
plt.ylabel('Number of Locations')
plt.title('Top 10 Fast Food Chains in USA')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

# Top 10 States with Most Restaurants


```python
top_states = df['province'].value_counts().head(10)
print("Top 10 states with most restaurants:")
top_states
```

```python
plt.figure(figsize=(10, 6))
top_states.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Number of Restaurants')
plt.title('Top 10 States by Number of Fast Food Restaurants')
plt.xticks(rotation=45)
plt.tight_layout()
```

# Top 10 Cities with Most Restaurants


```python
top_cities = df['city'].value_counts().head(10)
print("Top 10 cities with most restaurants:")
top_cities
```

```python
plt.figure(figsize=(12, 6))
top_cities.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.title('Top 10 Cities by Number of Fast Food Restaurants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

# Popular Restaurant Chains Comparison


```python
chains_of_interest = ["McDonald's", "Subway", "Burger King", "Wendy's", "Taco Bell"]
chain_counts = df[df['name'].isin(chains_of_interest)]['name'].value_counts()
print("Popular chains comparison:")
chain_counts
```

```python
plt.figure(figsize=(10, 6))
chain_counts.plot(kind='bar')
plt.xlabel('Restaurant Chain')
plt.ylabel('Number of Locations')
plt.title('Comparison of Major Fast Food Chains')
plt.xticks(rotation=45)
plt.tight_layout()
```
